# Detection and Translation

```python
from langdetect import detect
from deep_translator import GoogleTranslator
```

```python
def detect_and_translate(text, target_lang):
    result_lang = detect(text)
    translator = GoogleTranslator(source='auto', target=target_lang)
    translate_text = translator.translate(text)
    return result_lang, translate_text
```

```python
#example
sentence = "We are living in the era of technology."
result_lang, translate_text = detect_and_translate(sentence, target_lang='fr')
print("Detected Language:", result_lang)
print("Translation:", translate_text)
```
